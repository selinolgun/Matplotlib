### Grafikleri statik halde görebilmek için komutu kullandık.

## Etiket ve Legend Ayarları

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
veri1 = np.random.randn(1000).cumsum()
veri2 = np.random.randn(1000).cumsum()
veri3 = np.random.randn(1000).cumsum()
ax.plot(veri1, "k", label= "bir")
ax.plot(veri2, "b--", label= "iki")
ax.plot(veri3, "r.", label= "uc")
ax.legend(loc= "best")
```

### Üç farklı verimizi grafikte ekrana yazdırdık. `legend()` metoduyla grafiğimizi oluştururken kullandığımız `label=` parametresiyle belirlediğimiz isimlerin açıklamasını grafiğimize ekler.
### `legend()` metodunun aldığı `loc=` parametresiyle bu açıklamanın yerini belirleyebiliriz. `best` olarak girdiğimizde ise matplotlib bizim için en uygun yeri belirler ve oraya koyar.
### Ayrıca `loc=` parametresinin ön tanımlı hali zaten `best` şeklindedir. Yani yazmasakta aynı etkiyi görürdük.

## Grafiğe Yazı ve Ok Ekleme

```python
import pandas as pd
```

### Bu kısımda kullanacağımız için Pandas kütüphanesini içe aktardık.

```python
veri = pd.read_csv("FB.csv")
veri.head()
```

### `read_csv()` metoduyla veri setimizi okuttuk.
### Nasıl bir veri setiyle karşı karşıya olduğumuzu anlamak için `head()` metoduyla ilk 5 satırımızı görüntüledik.

```python
veri = pd.read_csv("FB.csv", index_col= 0, parse_dates= True)
veri.head()
```

### Veri setimizde index numaraları yerine tarih değerlerinin bulunmasını istedik. Bu yüzden `index_col= 0` parametresiyle değerleri sıfırladık. `parse_dates= True` değerleriyle de Date sütununu index değerleri olarak aldık.

```python
k_fiyat = veri["Close"]
```

### `k_fiyat` değişkenine `Close` isimli sütunumuzu atadık.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
k_fiyat.plot(ax= ax, style= "k-")
ax.set_title("Facebook Borsa Kapanış Fiyatları")
```

### `figure()` metoduyla çalışacağımız alanı belirledik.
### `ax` değişkenine `add_subplot()` metoduyla bir grafik atadık.
### `plot()` metoduyla grafiğimizi oluşturduk.
### `set_title()` metoduyla da grafiğimize başlık ekledik.

```python
k_fiyat.min()
```

### Pandasta bulunan `min()` metoduyla en düşük değeri bulabiliyorduk.

```python
k_fiyat["2018-12-24"]
```

### İndex değerini girerekte aynı veriye ulaşabilmemiz mümkündü.

```python
tarih = "2018-12-24"
yazi = "Minimum Değer"
```

### Gerekli değişkenlerimizi atadık.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
k_fiyat.plot(ax= ax, style= "k-")
ax.set_title("Facebook Borsa Kapanış Fiyatları")
ax.annotate(yazi, xy= (tarih, 125), xytext= (tarih, 150), arrowprops= dict(facecolor= "red"))
plt.show()
```

### `annotate()` metoduyla grafiğimize ok ekleyebiliyoruz.
### İlk parametre olarak okumuzun adını belirlediğimiz değişkeni yazdık.
### İkinci parametre de `xy=` okumuzun kordinatlarını giriyoruz. x  yerine tarih yazarak tarihimizi atadığımız yere gelmesini sağladık. y kordinatıda kendi kararımızca verilebilir. Deneyerek görebilirsiniz.
### Üçüncü parametrede de `xytext=` okumuzun üstüne yazacağımız yazının koordinatını belirliyoruz.
### Dördüncü parametrede ise `arrowprops=` okumuzun rengini `dict(facecolor= "red")` kırmızı olarak belirliyoruz.

## Grafiği Kaydetme

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
k_fiyat.plot(ax= ax, style= "k-")
ax.set_title("Facebook Borsa Kapanış Fiyatları")
ax.annotate(yazi, xy= (tarih, 125), xytext= (tarih, 150), arrowprops= dict(facecolor= "red"))
plt.savefig("grafigim.pdf")
```

### `savefig()` metoduyla grafiğimizi istediğimiz formatta kaydedebiliriz.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
k_fiyat.plot(ax= ax, style= "k-")
ax.set_title("Facebook Borsa Kapanış Fiyatları")
ax.annotate(yazi, xy= (tarih, 125), xytext= (tarih, 150), arrowprops= dict(facecolor= "red"))
plt.savefig("grafigim.png", dpi= 500, bbox_inches= "tight")
```

### `savefig()` metodu kullanırken 2 önemli parametre daha kullanabiliriz.
### Bunlardan birincisi `dpi=` parametresidir. Bu parametreyle grafiğimizi kaydederken çözünürlüğünü belirleyebiliyoruz.
### İkincisi ise `bbox_inches=` parametresidir. Bu parametreye `tight` yazarsak görselimiz kaydedilirken kenarında kalan boş kısımlar kesilerek kaydedilir.
