# Mathplotlib - 1

### Matplotlib Python'da veriyi görselleştirmek için kullanılan bir kütüphanedir.

### Matplotlib kütüphanesini yüklemek için `pip install matplotlib` komutunu terminalde kullanmanız gerekir.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Matplotlib kütüphanesini kullanırken `pyplot` modülünü alarak kullanıyoruz ve kısaltma olarakta `plt` kısaltmasından yararlanıyoruz.
### Matplotlib kütüphanesi sayısal verileri görselleştirirken Numpy kütüphanesini sık olarak kullanır. Bu yüzden ondan da yararlanmak için Numpy kütüphanesini de içe aktardık.

```python
veri = np.arange(10)
```

### Numpy kütüphanesinden yararlanarak hızlıca 10 elemanlı bir array oluşturduk.

```python
%matplotlib inline
```

### Bu komut ile çizeceğimiz grafikler ayrı bir pencere olarak karşımıza gelir.

```python
plt.plot(veri)
```

### Verimiz 0-10 arasındaki değerler olduğu için bu değerlerde 0' dan başlayarak 10' a kadar bir grafik çizdi. Bu grafiği oluşturmak için `plot()` metodunu kullandık.
### `plot()` metodu bizim çizgisel grafikler oluşturmamızı sağlar.

```python
plt.plot(veri)
plt.title("Başlık")
plt.xlabel("x Tarafı")
plt.ylabel("y Tarafı")
```

### `title()` metoduyla grafiğimize başlık, `xlabel()` metoduyla grafiğimizin x eksenine isim, `ylabel()` metoduyla grafiğimizin y eksenine isim oluşturabiliriz.

```python
%matplotlib notebook
```

### Bu komutu kullanarak grafiğimizi interaktif hale getirebiliriz. Ancak bu komut Visual Code programında çalışmıyor. Başka uygulama kullanıyorsanız deneyerek öğrenebilirsiniz.

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)
```

### `plot()` metodu kullanarak önceden tanımladığımız `x` ve `y` değerlerimizi grafik haline getirebiliyoruz. 
### Burada önceden tanımlanmış değerlerin eşit eleman sayısına sahip olması önemli bir detaydır.

```python
plt.plot(x, y, "ro")
```

### `plot()` metodunda bir parametre daha vardır.
### Bu parametre ön tanımlı olarak b- yani blue(b) ve line(-)' dır.
### Bu yüzden eğer belirtmezsek grafiğimiz mavi çizgi şekilde gösterilir.
### Biz bu parametreyi `ro` ile değiştirdik.
### Bu yüzden grafiğimiz gördüğünüz gibi red(r) ve noktalardan(o) oluştu. 

```python
d = np.arange(0., 5., 0.2)
plt.plot(d, d, "r--", d, d**2, "bs", d, d**3, "g^")
```

### Bir grafikte birden fazla gösterimde oluşturabiliriz.
### Bu grafikte d' yi , d' nin karesini ve d' nin küpünü gösterdik.
### Daha önce görmediğimiz özelleştirmelerden `--` tireleri, `s` kareleri(square), `g` yeşili(green), `^` üçgenleri ifade ediyor.

```python
d = np.arange(0., 5., 0.2)
plt.plot(d, d, "r--", d, d**2, "bs", d, d**3, "g^")
plt.savefig("grafigim1.png")
```

### `savefig()` metoduyla oluşturduğumuz grafiği bilgisayarımıza kaydedebiliriz.
### Kaydettiği konum çalışma dosyamızın konumu ile aynı yerdir.
### Farklı formatlarda kaydetmek için dosya uzantısını değiştirebiliriz.

```python
d = np.arange(0., 5., 0.2)
plt.plot(d, d, "r--", d, d**2, "bs", d, d**3, "g^")
plt.savefig("grafigim2.png", dpi= 500, bbox_inches= "tight")
```

### `savefig()` metodu kullanırken 2 önemli parametre daha kullanabiliriz.
### Bunlardan birincisi `dpi=` parametresidir. Bu parametreyle grafiğimizi kaydederken çözünürlüğünü belirleyebiliyoruz.
### İkincisi ise `bbox_inches=` parametresidir. Bu parametreye `tight` yazarsak görselimiz kaydedilirken kenarında kalan boş kısımlar kesilerek kaydedilir.
